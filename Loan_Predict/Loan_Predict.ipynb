# Check eligibility of the Customer given the inputs described above.
Classification:

# Define X and Y

```python
Y=training[["Loan_Status"]]
X=training.drop(labels=["Loan_Status","Loan_ID"],axis=1)
```

# Preprocessing

```python
from NS6PM import seprate
cat,con=seprate(X)
```

```python
x1=X[con]
x2=pd.get_dummies(X[cat])
Xnew=x1.join(x2)
```

# Data Spliting

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(Xnew,Y,test_size=0.2,random_state=31)
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(random_state=21, n_estimators=20)
```

```python
rfr = RandomForestClassifier(random_state=21, n_estimators=20)
model=rfr.fit(xtrain,ytrain)

tr_pred=model.predict(xtrain)
ts_pred=model.predict(xtest)

from sklearn.metrics import accuracy_score
tr_acc=accuracy_score(ytrain,tr_pred)
ts_acc=accuracy_score(ytest,ts_pred)
print("TR_acc:",tr_acc)
print("TS_acc:",ts_acc)
```

```python
tr = []
ts = []
for i in range(2,30):
    rfc = RandomForestClassifier(random_state=21,n_estimators=i)
    model = rfc.fit(xtrain,ytrain)
    tr_pred = model.predict(xtrain)
    ts_pred = model.predict(xtest)
    from sklearn.metrics import accuracy_score
    tr_acc = accuracy_score(ytrain,tr_pred)
    ts_acc = accuracy_score(ytest,ts_pred)
    tr.append(tr_acc)
    ts.append(ts_acc)
```

```python
import matplotlib.pyplot as plt
plt.plot(tr,c="blue")
plt.plot(ts,c="red")
```

# GridSearchCV

```python
tg = {"n_estimators":range(2,20,1),"max_depth":range(2,20)}
from sklearn.ensemble import RandomForestClassifier
dtr = RandomForestClassifier(random_state=21)
from sklearn.model_selection import GridSearchCV
cv = GridSearchCV(dtr,tg,scoring="accuracy",cv=4)
cvmodel = cv.fit(Xnew,Y)
cvmodel.best_params_
```

```python
from sklearn.ensemble import RandomForestClassifier
rfr = RandomForestClassifier(random_state=21, n_estimators=17,max_depth=5)
model=rfr.fit(xtrain,ytrain)

tr_pred=model.predict(xtrain)
ts_pred=model.predict(xtest)

from sklearn.metrics import accuracy_score,confusion_matrix,recall_score
tr_acc=accuracy_score(ytrain,tr_pred)
ts_acc=accuracy_score(ytest,ts_pred)
print("TR_acc:",tr_acc)
print("TS_acc:",ts_acc)
```
